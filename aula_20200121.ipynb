```
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```
xs = np.array([1,2,3,4,5], dtype=np.float64)
ys = np.array([5,4,6,5,6], dtype=np.float64)
plt.scatter(xs,ys,color='#003F72')
plt.show()
```

A intenção nesse exercício é calcular a linha de melhor ajuste entre os pontos. Para isso utilizaremos as formulas a seguir.

```
def calcular_pontos(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)**2) - mean(xs**2)))
    
    b = mean(ys) - m*mean(xs)
    
    return m, b

m, b = calcular_pontos(xs,ys)
line = [(m*x)+b for x in xs]

plt.scatter(xs,ys,color='#003F72')
plt.plot(xs, line)
plt.show()
```

```
predict_x = 5.7
predict_y = (m*predict_x)+b

plt.scatter(xs,ys,color='#003F72',label='data')
plt.scatter(predict_x,predict_y,color='#3F0072',label='prediction')
plt.plot(xs, line, label='regression line')
plt.legend(loc=4)
plt.show()
```
